##Plot 꾸미기

###Line Plot

```
plt.plot(np.random.randn(30), color = "g", marker = "o", linestyle = "--")
```

```
#컬러, 마킹, 라인스타일
plt.plot(np.random.randn(40), "k.-")
```

###bar Plot

```
fig2, axes2 = plt.subplots(2,1)
```

```
data = pd.Series(np.random.randn(16), index = list("abcdefghijklmnop"))
```

```
data
```

```
axes2
```

```
#alpha == 투명도
data.plot(kind = "bar", ax = axes2[0], color = 'k', alpha = 0.7)
```

```
fig2
```

```
data.plot(kind = "barh", ax=axes2[1], color ='g', alpha = 1)
```

```
fig2
```

##눈금 설정

```
fig3 = plt.figure()
```

```
ax = fig3.add_subplot(1,1,1)
```

```
fig3
```

```
ax.plot(np.random.randn(1000).cumsum())
```

```
fig3
```

x축 - xtick, y축 - ytick

```
ax.set_xticks([0,250,500,750,1000])
```

```
fig3
```

```
ax.set_yticks([10,0,-10])
```

```
fig3
```

```
ax.set_xticklabels(["one", "two", "three", "four", "five"],
                            rotation = 30, fontsize = "small")
```

```
fig3
```

##제목 설정

```
ax.set_title("Random Plot")
```

```
fig3
```

```
 ax.set_xlabel("Stages")
 ax.set_ylabel("Values")
```

```
fig3
```

##범례 설정

```
fig4 = plt.figure()
ax4 = fig4.add_subplot(1,1,1)
```

```
fig4
```

```
ax4.plot(np.random.randn(1000).cumsum(), 'k', label = "one")
ax4.plot(np.random.randn(1000).cumsum(),  "b--", label = "two")
ax4.plot(np.random.randn(1000).cumsum(), "r.", label = "three")
```

```
fig4
```

```
#best = 최적의 위치
ax4.legend(loc="best")
```

```
fig4
```
